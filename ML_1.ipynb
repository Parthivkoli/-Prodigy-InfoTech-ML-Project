```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset (Assuming the CSV file is in the correct format)
file_path = r'housing_price_dataset.csv'
data = pd.read_csv(file_path)

# Select relevant features and the target variable
# Adjust column names based on the dataset you've described
numeric_data = data[['SquareFeet', 'Bedrooms', 'Bathrooms', 'Price']]

# Display correlation heatmap to understand the relationship between features
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Select the features (square footage, number of bedrooms, and number of bathrooms) and target variable (price)
features = data[['SquareFeet', 'Bedrooms', 'Bathrooms']]
target = data['Price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model using Mean Squared Error (MSE) and R-squared (R²) score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")

# Display the coefficients of the model to see the impact of each feature
coefficients = pd.DataFrame(model.coef_, features.columns, columns=['Coefficient'])
print("\nModel Coefficients:")
print(coefficients)
```
